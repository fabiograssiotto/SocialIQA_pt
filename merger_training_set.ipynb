### Imports

```python
import os
import pandas as pd
```

### Merge training dataset results (PT)

```python
# List of CSV file paths
folder_path = "translated/final/train_sets"
file_paths = ['train_1.csv', 'train_2.csv', 'train_3.csv', 'train_4.csv', 'train_5.csv', 'train_6.csv']#, 'train_7.csv'] # Add as needed

# Initialize an empty list to store the DataFrames
dataframes = []

# Read each CSV file and append to the list
for file in file_paths:
    df = pd.read_csv(folder_path + "/" + file)
    dataframes.append(df)

# Concatenate the DataFrames row-wise
df_train = pd.concat(dataframes, ignore_index=True)
df_train = df_train.drop(columns=['Unnamed: 0'])
```

```python
df_train
```

```python
# Write to filesystem
df_train.to_csv("translated/final/train.csv") 
```

### Merge Ranking results

```python
# List of CSV file paths
folder_path = "rankings/train_sets"
file_paths = ['train_1_rankings.csv', 'train_2_rankings.csv', 'train_3_rankings.csv',
              'train_4_rankings.csv', 'train_5_rankings.csv', 'train_6_rankings.csv']#, 'train_7_rankings.csv'] # Add as needed

# Initialize an empty list to store the DataFrames
dataframes = []

# Read each CSV file and append to the list
for file in file_paths:
    df = pd.read_csv(folder_path + "/" + file)
    dataframes.append(df)

# Concatenate the DataFrames row-wise
df_train_rankings = pd.concat(dataframes, ignore_index=True)
df_train_rankings = df_train_rankings.drop(columns=['Unnamed: 0'])
```

```python
# Write to filesystem
df_train_rankings.to_csv("rankings/train_rankings.csv") 
```
